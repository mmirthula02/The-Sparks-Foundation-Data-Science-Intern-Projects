### Objective : To classify the species using decision tree regressor 

1. Loading the necessary iris datasets

```
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
```

2. Seperating the feature and target variable

```
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)
X.head()
```

```
y = iris.target
y
```

3. Training the Decision tree model

```
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
regressor = DecisionTreeRegressor()

regressor.fit(X_train,y_train)
print("Model successfully trained!!")
```

4. Predicting the accuracy score

```
y_pred = regressor.predict(X_test)

accuracy = regressor.score(X_test,y_test)
print("Model accuracy {}".format(accuracy))
```

5. Vizualizing the graph using graphviz module 

```
from sklearn import tree
import graphviz 

dot_data = tree.export_graphviz(regressor,out_file=None,filled=True,rounded=True,feature_names=iris.feature_names,  
                                class_names=iris.target_names)

graph=graphviz.Source(dot_data)
graph
```

6. Accuracy vs Prediction

```
df= pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
df.head(10)
```

7. Providing new data

```
new_data= [[3.5,3,1.5,0.2]]
new_predic = regressor.predict(new_data)
print("Classification: {}".format(new_predic[0]))
```

Vizualizing graph through plot module

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,10))
plot = tree.plot_tree(regressor, feature_names=iris.feature_names, class_names=iris.target_names,  filled=True)
plot
```
